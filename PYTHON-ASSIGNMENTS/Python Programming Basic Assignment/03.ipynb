```python
# 1. Write a Python Program to Check if a Number is Positive, Negative or Zero?
inp1 = int(input("Enter a number: "))
if inp1 > 0:
    print(f"{inp1} is a positive integer")
elif inp1 == 0:
    print(f"{inp1} is a ZERO.")
else:
    print(f"{inp1} is a negative integer")
```

```python
# 2. Write a Python Program to Check if a Number is Odd or Even?

inp1 = int(input("Enter a number: "))
if inp1%2 == 0:
    print(f"{inp1} is a even number.")
else:
    print(f"{inp1} is a odd number")
```

```python
# 3. Write a Python Program to Check Leap Year?
y = int(input("Enter a year: "))
if y%4 != 0:
    print("NOT A LEAP YEAR")
elif y%100 != 0:
    print("LEAP YEAR")
elif y%400 != 0:
    print("NOT A LEAP YEAR")
else:
    print("LEAP YEAR")
```

```python
# 4. Write a Python Program to Check Prime Number?

def check_prime(inp1):
    if inp1 == 0 or inp1 == 1:
        return False
    elif inp1 == 2 or inp1 == 3:
        return True
    elif inp1%2 == 0 or inp1%3 == 0:
        return False
    else:
        i=5
        while i*i<=inp1:
            if inp1%i == 0 or inp1%(i+2) == 0:
                return False
            i+=6
        return True

inp1 = int(input("Enter a number: "))
res = check_prime(inp1)
if res == True:
    print(f"{inp1} is a prime number")
else:
    print(f"{inp1} is not a prime number")
```

```python
# 5. Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?
def check_prime(inp1):
    if inp1 == 0 or inp1 == 1:
        return False
    elif inp1 == 2 or inp1 == 3:
        return inp1
    elif inp1%2 == 0 or inp1%3 == 0:
        return False
    else:
        i=5
        while i*i<=inp1:
            if inp1%i == 0 or inp1%(i+2) == 0:
                return False
            i+=6
        return inp1
    
s = int(input("Enter the range: [start]"))
e = int(input("Enter the range: [end]"))
for i in range(s, e+1):
    res = check_prime(i)
    if res > 0:
        print(f"{res} ", end=" ")
```
